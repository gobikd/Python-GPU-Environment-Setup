```python
############## Section 1 – System Information ##############
import platform
import sys
import os

print("="*60)
print("SYSTEM INFORMATION")
print("="*60)

print("Operating System :", platform.platform())
print("Python Version   :", platform.python_version())
print("Python Executable:", sys.executable)
print("Current Directory:", os.getcwd())
```

```python
############## Section 2 – Library Versions ##############
import numpy as np
import pandas as pd
import matplotlib
import sklearn

print("="*60)
print("LIBRARY VERSIONS")
print("="*60)

print("NumPy        :", np.__version__)
print("Pandas       :", pd.__version__)
print("Matplotlib   :", matplotlib.__version__)
print("Scikit-Learn :", sklearn.__version__)
```

```python
############## PyTorch Information ##############
import torch

print("="*60)
print("PYTORCH")
print("="*60)

print("Torch Version :", torch.__version__)
print("CUDA Available:", torch.cuda.is_available())
print("CUDA Runtime  :", torch.version.cuda)
print("cuDNN Enabled :", torch.backends.cudnn.enabled)
print("cuDNN Version :", torch.backends.cudnn.version())
```

```python
############## GPU Information ##############
print("="*60)
print("GPU INFORMATION")
print("="*60)

if torch.cuda.is_available():
    print("GPU Name      :", torch.cuda.get_device_name(0))
    print("GPU Memory GB :", round(
        torch.cuda.get_device_properties(0).total_memory / 1024**3,2))
```

```python
############## Matrix Multiplication Benchmark ##############
import time

device = "cuda"

x = torch.rand(5000,5000,device=device)
y = torch.rand(5000,5000,device=device)

torch.cuda.synchronize()

start = time.time()

z = x @ y

torch.cuda.synchronize()

end = time.time()

print("="*60)
print("GPU BENCHMARK")
print("="*60)

print("Time :", round(end-start,3),"seconds")
print("Device :", z.device)
```

```python
############## CNN Test ##############
import torch.nn as nn

model = nn.Conv2d(
    in_channels=3,
    out_channels=16,
    kernel_size=3,
    padding=1
).to(device)

x = torch.randn(8,3,224,224).to(device)

with torch.no_grad():
    y = model(x)

print("="*60)
print("CNN TEST")
print("="*60)

print("Output Shape :", y.shape)
print("Device :", y.device)
```
